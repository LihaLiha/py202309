```python
import random # 랜덤 함수 가져오기



def get_number():
    return random.randrange(1, 46)  # 랜덤으로 범위는 1 ~ 46

def set_lotto_numbers():
    
    lotto = [] 
    num = 0


    # 6개의 숫자가 나올때까지 반복
    while True:
        num = get_number()
    
        if lotto.count(num) == 0:
            lotto.append(num)
        
        if len(lotto) >= 6:
            break
    return lotto
    
    
# 로또 숫자 비교 함수
def match_numbers(user_input, lotto_numbers):
    
    same_numbers = set(user_inputs)&set(lotto_numbers)
    
    return same_numbers

# 맞은 개수에 따른 등수 함수
def check_prize(same_numbers):
    
    if len(same_numbers) == 6:
        print("first place")
    elif len(same_numbers) == 5:
        print("second place")
    elif len(same_numbers) == 4:
        print("third place")
    elif len(same_numbers) == 3:
        print("fourth place")
    else:
        print("no prize")

        
num = 0
print("** Start Lotto **")
print("1) Set the lotto numbers") # 로또 번호 입력받기
lotto_numbers = set_lotto_numbers()

print("2) Set the user inputs")

user_inputs = []

while True:
    num = int(input("Enter %d-number: "% (len(user_inputs) + 1)))
    if num < 1 or num > 46 :
        print("No valid number")
        continue
    if user_inputs.count(num) == 0:
        user_inputs.append(num)
    else:
        print("Same number")
        continue
    if len(user_inputs) >= 6:
        break
        
print("3) Match the numbers") # 로또 번호 비교 함수 호출
same_numbers = match_numbers(user_inputs, lotto_numbers)

print("4) Check the prize") # 등수 함수 호출
check_prize(same_numbers)

print("5) Sort the lists") # 번호 오름차순 배열
lotto_numbers.sort()
user_inputs.sort()

# 결과 출력
print("=== Summary ===")
print("Lotto numbers: ", *lotto_numbers)
print("User numbers: ", *user_inputs)

```
